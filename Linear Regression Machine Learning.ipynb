```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
try:
    df = load_boston()
    print(df)
except: 
    print("successfully done!")
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset
```

```python
dataset.head()
```

```python
X=dataset
y=df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( 
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train) 
```

```python
X_train
```

```python
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring=
                    "neg_mean_squared_error",cv=7)
```

```python
mse
```

```python
np.mean(mse)
```

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
```

```python
sns.distplot(reg_pred-y_test,kde=True,color='green')
```

```python
slope = regression.coef_[0]
slope
```

```python
plt.scatter(y_test, reg_pred, c = 'red')
plt.xlabel("Price")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")

plt.show()
```

```python
plt.scatter(y_test, reg_pred, c = 'green')
plt.xlabel("Price")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.plot([min(y_test), max(y_test)], 
         [min(y_test) * slope, max(y_test) * slope], 
         color='blue', linestyle='-', linewidth=2)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test,reg_pred)
mae
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_pred,y_test)
```
