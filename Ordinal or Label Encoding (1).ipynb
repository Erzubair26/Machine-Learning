```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

```python
d = {
    "Education": ["PHd", "UG", "PG", "SCHOOL"],
    "Review": ["Average", "Poor", "Good", "Excellent"],
    "Result": ["YES", "NO", "NO", "YES"]
    }
```

```python
d
```

```python
df=pd.DataFrame(d)
```

```python
df
```

```python
od_encoder = OrdinalEncoder(categories=[["SCHOOL", "UG", "PG", "PHd"],["Poor", "Average", "Good", "Excellent"]])
```

```python
x = ["Education", "Review"]
```

```python
df[x] = od_encoder.fit_transform(df[x])
```

```python
print(df)  
```

```python
le_result = LabelEncoder()
df['Result'] = le_result.fit_transform(df['Result'])
```
